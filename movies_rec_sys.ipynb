```python
#need to extract genres from list of dicts
import ast
def fun(strlistdic):
    L = []
    strlistdic = ast.literal_eval(strlistdic)
    for items in strlistdic:
        #items is a dict
        L.append(items['name'])
    return L

def fun2(strlistdic):
    L = []
    strlistdic = ast.literal_eval(strlistdic)
    for items in strlistdic:
        #items is a dict
        L.append(items['name'])
    return L[:3]

def fun3(strlistdic):
    L = []
    strlistdic = ast.literal_eval(strlistdic)
    for items in strlistdic:
        if items['job']=='Director':
            L.append(items['name'])
            break
    return L

ds2['genres']=ds2['genres'].apply(fun)
ds2['keywords']=ds2['keywords'].apply(fun)
ds2['cast']=ds2['cast'].apply(fun2)
ds2['crew']=ds2['crew'].apply(fun3)
ds2['overview']  = ds2['overview'].apply(lambda x : x.split())

```

```python
ds2['genres']=ds2['genres'].apply(lambda x:[i.replace(" ","") for i in x])
ds2['keywords']=ds2['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
ds2['cast']=ds2['cast'].apply(lambda x:[i.replace(" ","") for i in x])
ds2['crew']=ds2['crew'].apply(lambda x:[i.replace(" ","") for i in x])
ds2['overview']  = ds2['overview'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
ds2['answer'] = ds2['genres'] + ds2['keywords'] + ds2['cast'] + ds2['crew'] + ds2['overview']
```

```python
ds2_new = ds2[['answer', 'id', 'title']]
```

```python
ds2_new['answer']= ds2_new['answer'].apply(lambda x : " ".join(x))
```

```python
# Lowercase the text in 'answer' column
ds2_new['answer'] = ds2_new['answer'].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def fun4(txt):
    lst = []
    for i in txt.split():
        lst.append(ps.stem(i))

    return " ".join(lst)

# Apply stemming and save the result back to the DataFrame
ds2_new['answer'] = ds2_new['answer'].apply(fun4)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
countvec = CountVectorizer(max_features=6900, stop_words='english')
movie_vector = countvec.fit_transform(ds2_new['answer']).toarray()

movie_vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim_Scr = cosine_similarity(movie_vector)
```

```python
def rec(movie):
    idx = ds2_new[ds2_new['title']==movie].index[0]
    top5 = sorted(list(enumerate(sim_Scr[idx])),reverse=True, key = lambda x:x[1])
    top5 = top5[1:6]
    l2 = []
    for items in top5:
        l2.append(ds2_new.iloc[items[0]].title)
    return l2

rec('Titanic')


    
```
